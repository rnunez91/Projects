# Questions Chosen
##### Q1 = I traded without considering the information pieces provided to me
##### Q2 = I traded using ALL of the information provided to me at each stage
##### Q3 = I traded using SOME of the information provided to me at each stage 
##### Q8 = I did not want the cash rewards at all
##### Q30 = I have had prior experience in live trading stock /shares

```python
survey_data['Q1'] = survey_data['Q1'].astype('category')
survey_data['Q2'] = survey_data['Q2'].astype('category')
survey_data['Q3'] = survey_data['Q3'].astype('category')
survey_data['Q8'] = survey_data['Q8'].astype('category')
survey_data['Q30'] = survey_data['Q30'].astype('category')
survey_data['Gender'] = survey_data['Gender'].astype('category')

survey_data.dtypes
```

```python
cat_columns = survey_data.select_dtypes(['category']).columns
cat_columns
```

```python
survey_data[cat_columns] = survey_data[cat_columns].apply(lambda x: x.cat.codes)

survey_data
```

##### Q1. Interestingly it appears that both genders traded without reviewing information roughly the same though more men answered neutrally.
##### Q2. More women strongly agreed to using all information given but also strongly disagreed compared to the men.
##### Q3. Though the general consensus was a "agree", more men a disagreed to use some of the information while more women strongly agreed.

##### Q30. More women had less experience than men in trading stocks/shares.
